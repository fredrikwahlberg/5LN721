# Plotting script for colour intensities in pixels

```python
import cv2
import numpy as np

import matplotlib.pyplot as plt
```

```python
I = cv2.cvtColor(cv2.imread("/content/Apricot_flowering_2021_crop.jpg"), cv2.COLOR_BGR2RGB)

plt.figure()
plt.imshow(I)
plt.axis('off')
plt.show()
```

```python
fig = plt.figure(figsize=(24, 8))
ax = fig.subplots(1, 3)
for i in range(3):
  J = np.zeros(I.shape, dtype=I.dtype)
  J[:, :, i] = I[:, :, i]
  ax[i].imshow(J)
  ax[i].axis('equal')
  ax[i].axis('off')
fig.tight_layout(pad=0)
fig.show()
```

```python
fig = plt.figure(figsize=(24, 8))
ax = fig.subplots(1, 3)
for i in range(3):
  J = np.zeros(I.shape, dtype=I.dtype)
  J[:, :, i] = I[:, :, i]
  ax[i].imshow(J)
  for j in range(I.shape[0]):
    for k in range(I.shape[1]):
      # ax[i].text(j, k, "%s" % I[k, j, i], c='lightgray', fontsize=8, horizontalalignment='center', verticalalignment='center')
      ax[i].text(j, k, "%s" % I[k, j, i], c='w', alpha=.9, fontsize=8, horizontalalignment='center', verticalalignment='center')
  ax[i].axis('equal')
  ax[i].axis('off')
fig.tight_layout(pad=0)
fig.show()
```
